Queremos generar variables aleatorias según la distribución exponencial con parámetro $\lambda=1$

La PDF de la distribución exponencial es $f(x)=e^{-x}$ y $g(x)=1$ puesto que es facil trabajar con una distribución uniforme


```python
import random as rnd
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def aceptacion_rechazo(n):
    muestras = []
    while len(muestras) < n:
        X = rnd.random()  # Genera X ~ Uniforme(0, 1)
        Y = rnd.random()  # Genera Y ~ Uniforme(0, 1)
        if Y <= math.exp(-X):  # Condicion de aceptacion
            muestras.append(X)
    return muestras
```

```python
H = aceptacion_rechazo(1000)
```

```python
plt.hist(H,30)
```

```python
plt.plot(H)
```

```python
def cfd(muestra):
  n = len(muestra)
  x = np.sort(muestra)
  y = np.arange(0,n)/n
  return x,y

x,y = cfd(H)
```

```python
plt.step(x,y)
```
