set() crea un conjunto vacío, y values_seen.add(x1) agrega el valor x1 al conjunto values_seen.
Estos conjuntos son utiles cuando queremos almacenar elementos unicos.

El bucle 'while true' continuará ejecutándose indefinidamente hasta que se cumpla la condición de salida.

```python
#Metodo de congruencia
#a)

x0 = 21
a = 21
c = 15
m = 31

r = []
values_seen = set()
iteracion = 0

while True:
    x1 = (a * x0 + c) % m

    if x1 in values_seen:
        break

    values_seen.add(x1)
    r.append(x1 / m)

    x0 = x1
    iteracion += 1

print('Iteracion:',iteracion)
print(r)
```
