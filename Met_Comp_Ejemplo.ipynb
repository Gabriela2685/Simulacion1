a=8 b=9 c=10

Obtenemos las áreas coorespondientes:

$A_1=\frac{9-8}{10-8}=\frac{1}{2}$

$A_2=\frac{(10-9)}{10-8}=\frac{1}{2}$

De la primer parte de la función tenemos las sig. coordenadas $p_{1}(8,\frac{1}{4})$ y $p_{2}=(9,\frac{3}{4})$

$f_{1}=\frac{1}{2}(x-8)+4$

De la segunda parte de la función tenemos las sig. coordenadas $p_{2}=(9,\frac{3}{4})$ y $p_{3}(10,\frac{1}{4})$

$f_{2}=-\frac{1}{2}(x-9)+\frac{3}{4}$

Entonces la función de probabilidad queda como:

$\left\{ \begin{array}{lcc} \frac{1}{2}(x-8)+4 & si & 8 \leq x \leq 2 \\ \\ -\frac{1}{2}(x-9)+\frac{3}{4} & si & 9 \leq x \leq 10 \end{array} \right.$

Quedando la función de probabilidad como:

$f(x)=\frac{1}{2}(\frac{1}{2}(x-8)+4)+\frac{1}{2}(-\frac{1}{2}(x-9)+\frac{3}{4})=(\frac{1}{4}(x-8)+2)+(\frac{1}{4}(x-9)+\frac{3}{4})$

Obtenemos la función acumulada de las áreas: $\int_{∞}^{x}f(t)dt$

$\left\{ \begin{array}{lcc} \frac{(x-\frac{15}{2})^2}{4}-\frac{1}{16} & si & 8 \leq x \leq 2 \\ \\ \frac{1}{4}(x-\frac{21}{2})^2-\frac{9}{16} & si & 9 \leq x \leq 10 \end{array} \right.$

Igualamos a un $R_{2}$

$F_{1}=R_{2}$

$x=\sqrt{4R_{2}+\frac{1}{4}}+\frac{15}{2}$

$F_{2}=R_{2}$

$x=\sqrt{4R_{2}+\frac{9}{4}}+\frac{21}{2}$

```python
import sympy as sp
import numpy as np
import random as rnd
import matplotlib.pyplot as plt
from math import sqrt
```

```python
def comp(a, b, c, n):  # Creamos una función  que programe el ejercicio anterior
  lista = []           # En una lista agrgaremos los valores que se vayan generando
  A_1 = (b-a)/(c-a)
  for i in range(n):
    R_1 = rnd.random()   # Variables aleatorias
    R_2 = rnd.random()
    if R_1 < A_1:
      x = sqrt((4*R_2) + (1/4)) + (15/2)  # Inversa obtenida en el primer trozo
    else:
      x = sqrt((4*R_2) + (9/4)) + (21/2)   # Inversa obtenida en el segundo
    lista.append(x)
  return lista
```

```python
G = comp(8,9,10,1000)  # Damos los valores y el numero de iteraciones
```

```python
plt.hist(G,30)  # Histograma
```

```python
plt.plot(G)
```

```python
def cfd(muestra):
  n = len(muestra)
  x = np.sort(muestra)
  y = np.arange(0,n)/n
  return x,y

x,y = cfd(G)
```

```python
plt.step(x,y)
plt.title('Función de Distribución Acumulativa (CDF)')
```
