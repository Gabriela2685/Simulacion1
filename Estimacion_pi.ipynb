```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calcular_pi_aproximado(iteraciones):
    contador_puntos = 0
    lista_estimaciones = []
    for iteracion in range(iteraciones):
        punto_x = rnd.random()
        punto_y = rnd.random()
        if punto_y <= np.sqrt(1 - punto_x**2):
            contador_puntos += 1
            estimacion_actual = 4 * (contador_puntos / (iteracion + 1))
            lista_estimaciones.append(estimacion_actual)
    proporcion_final = contador_puntos / iteraciones
    pi_estimado = 4 * proporcion_final
    return lista_estimaciones, pi_estimado
```

```python
F, h=calcular_pi_aproximado(26000)
print(h)

plt.plot(F, color='blue', linestyle='-', linewidth=1.5, label='Estimación de pi')
plt.axhline(y=np.pi, color='red', linestyle='--', linewidth=1.5, label='Valor real de pi')
plt.axhline(y=3.14, color='green', linestyle='-.', linewidth=1, label='Aproximación común de pi')
plt.ylabel('Valor estimado de pi', fontsize=14)
plt.title('Estimación de pi', fontsize=16, fontweight='bold')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```
