**Método de la transformada inversa (Ejemplo)**

Proponemos FDA:

$f(x)= \left\{ \begin{array}{lcc} 0 & si & x < 0 \\ \\ \frac{x^2}{4} & si & 0 \leq x \leq 2 \\ \\ 1 & si & x > 2 \end{array} \right.$

Encontramos la inversa:

$\frac{x^2}{4} = U$

$x = \sqrt{U4}$

$x = 2 \sqrt{U}$ para $0 \leq x \leq 1$

```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Funcion de distribucion acumulada FDA

def FDA(n):    # Genera una lista de n números aleatorios que siguen la distribución definida por la FDA una n cantidad de veces
  lista = []
  for i in range(n):
    u = rnd.random()
    x = 2*np.sqrt(u)
    lista.append(x)
  return lista
```

```python
F = FDA(1000) # Asignamos valores a n
```

```python
plt.plot(F)
```

```python
plt.hist(F) # Muestra el histograma de cómo se distribuyen los números generados
            # Esperando una mayor concentracion cerca de 2
```

```python
def cfd(muestra):
  n = len(muestra)
  x = np.sort(muestra)
  y = np.arange(0,n)/n
  return x,y

x,y = cfd(F)
```

```python
plt.step(x,y) # Grafica de la acumulativa
```
