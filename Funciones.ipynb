Función de densidad

```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def expo(landa,n):
  L = -1/landa
  lista = []
  for i in range(n):
    x = rnd.random()
    U = L*np.log(x)
    lista.append(U)
  return lista
```

```python
G = expo(0.5,1000)
media = np.mean(G)
varianza = np.var(G)
```

Distribución acumulativa

```python
def cfd(muestra):
  n = len(muestra)
  x = np.sort(muestra)
  y = np.arange(0,n)/n
  return x,y

x,y = cfd(G)
```

```python
plt.hist(G)
```

```python
plt.plot(G)
```

```python
plt.step(x,y)
```

```python
def uniform(a,b,n):
  lista = []
  for i in range(n):
    x = rnd.random()
    U = (b-a)*x+a
    lista.append(U)
  return lista
```

```python
F = uniform(10,15,1000)
```

```python
x,y = cfd(F)
```

```python
plt.hist(F, color='orange')
```

```python
plt.plot(F, color='orange')
```

```python
plt.step(x,y, color='orange')
```

```python
def cauchy(x0,gamma,n):
  lista = []
  for i in range(n):
    x = rnd.random()
    U = gamma*np.tan((x-1/2)*np.pi)+x0
    lista.append(U)
  return lista
```

```python
H = cauchy(2,1,100)
```

```python
x,y = cfd(H)
```

```python
plt.hist(H, color='green')
```

```python
plt.plot(H, color='green')
```

```python
plt.step(x,y, color='green')
```

```python
def aprox(n):
  lista = []
  for i in range(n):
    x = rnd.random()
    U = (x**0.135-(1-x)**0.135)/0.1975
    lista.append(U)
  return lista
```

```python
M = aprox(10000)
```

```python
# Densidad
plt.hist(M, 50)
```

```python
plt.plot(M)
```

```python
# Acumulativa
x,y = cfd(M)
plt.step(x,y)
```

```python
def we(alpha,n):
  lista = []
  for i in range(n):
    x = rnd.random()
    U = (-np.log(x))**(1/alpha)
    lista.append(U)
  return lista
```

```python
J = we(1,1000)
plt.hist(J, 50)
```

```python
x,y = cfd(J)
plt.step(x,y)
```

```python
def Gf(n):
  lista = []
  for i in range(n):
    x = rnd.random()
    U = (-np.log(-np.log(x)))
    lista.append(U)
  return lista
```

```python
P = Gf(10000)
plt.hist(P,50)
```

```python
def triangular(a,b,c,n):
  C = []
  d1 = (b-a)
  d2 = (c-a)
  d3 = (c-b)

  for i in range(n):
    x = rnd.random()
    if x<=0:
      U = 0
      C.append(U)
    elif 0<x<(d1/d2):
      U = np.sqrt(d1*d2*x)+a
      C.append(U)
    elif d1/d2<x<1:
       k = -(x-1)
       U = -np.sqrt(k*d2*d3)+c
       C.append(U)
    elif x>=1:
      U = 1
      C.append(U)
  return C
```

```python
Y = triangular(100,150,200,1000)
```

```python
plt.hist(Y)
```
