Función de densidad

```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def expo(landa,n):
  L = -1/landa
  lista = []
  for i in range(n):
    x = rnd.random()
    U = L*np.log(x)
    lista.append(U)
  return lista
```

```python
G = expo(0.5,1000)
media = np.mean(G)
varianza = np.var(G)
```

Distribución acumulativa

```python
def cfd(muestra):
  n = len(muestra)
  x = np.sort(muestra)
  y = np.arange(0,n)/n
  return x,y

x,y = cfd(G)
```

```python
plt.hist(G)
```

```python
plt.plot(G)
```

```python
plt.step(x,y)
```

```python
def uniform(a,b,n):
  lista = []
  for i in range(n):
    x = rnd.random()
    U = (b-a)*x+a
    lista.append(U)
  return lista
```

```python
F = uniform(10,15,1000)
```

```python
x,y = cfd(F)
```

```python
plt.hist(F, color='orange')
```

```python
plt.plot(F, color='orange')
```

```python
plt.step(x,y, color='orange')
```

```python
def cauchy(x0,gamma,n):
  lista = []
  for i in range(n):
    x = rnd.random()
    U = gamma*np.tan((x-1/2)*np.pi)+x0
    lista.append(U)
  return lista
```

```python
H = cauchy(2,1,100)
```

```python
x,y = cfd(H)
```

```python
plt.hist(H, color='green')
```

```python
plt.plot(H, color='green')
```

```python
plt.step(x,y, color='green')
```
